# Import Required Libraries
Use Python libraries for HTML parsing and file handling.

```python
from pathlib import Path
try:
    from bs4 import BeautifulSoup
except ImportError:
    import sys, subprocess
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', 'beautifulsoup4'])
    from bs4 import BeautifulSoup
workspace_root = Path('d:/agntix/Intactdia Studio')
```

# Load HTML File Content
Read all HTML files in the workspace for processing.

```python
html_files = sorted(workspace_root.rglob('*.html'))
print(f'Found {len(html_files)} HTML files')
for html_file in html_files:
    print(html_file)
```

# Parse HTML and Add loading="lazy" to img Tags
Update every <img> tag to include lazy loading when not already present.

```python
summary = []
for html_file in html_files:
    text = html_file.read_text(encoding='utf-8')
    soup = BeautifulSoup(text, 'html.parser')
    changed = False
    for img in soup.find_all('img'):
        if not img.has_attr('loading'):
            img['loading'] = 'lazy'
            changed = True
    if changed:
        html_file.write_text(str(soup), encoding='utf-8')
        summary.append(str(html_file))
print(f'Updated {len(summary)} HTML files with lazy loading')
for file in summary:
    print(file)
```

# Analyze HTML for Performance Issues
Scan HTML files for duplicate CSS/JS includes and render-blocking resources.

```python
issues = []
for html_file in html_files:
    text = html_file.read_text(encoding='utf-8')
    if text.count('<link rel="stylesheet"') > 1:
        issues.append((str(html_file), 'possible duplicate stylesheet includes', text.count('<link rel="stylesheet"')) )
    if 'assets/js/three.js' in text:
        issues.append((str(html_file), 'heavy 3D script present: three.js'))
    if 'assets/js/hover-effect.umd.js' in text:
        issues.append((str(html_file), 'hover effect script may impact performance'))
print(f'Found {len(issues)} potential performance issues')
for issue in issues[:20]:
    print(issue)
```

# Optimize Resource Loading
Identify non-critical scripts that can use defer or async to improve page load.

```python
optimized = []
for html_file in html_files:
    text = html_file.read_text(encoding='utf-8')
    orig = text
    text = text.replace('<script src="assets/js/three.js"></script>', '<script defer src="assets/js/three.js"></script>')
    text = text.replace('<script src="assets/js/hover-effect.umd.js"></script>', '<script defer src="assets/js/hover-effect.umd.js"></script>')
    if text != orig:
        html_file.write_text(text, encoding='utf-8')
        optimized.append(str(html_file))
print(f'Applied deferred loading to {len(optimized)} HTML files')
for file in optimized:
    print(file)
```
